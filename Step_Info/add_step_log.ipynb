```python
import csv
stepList = list()
with open('Stepifier_Combined_Data.csv', newline='', encoding='utf-8') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',',
                            quotechar='|')
    for row in spamreader:
        print(row)
        if (len(row[0]) > 0):
            stepList.append(row)
csv.reader
```
