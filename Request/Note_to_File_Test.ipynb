```python
import requests
from atlassian import Confluence
import time
from getpass import getpass

contentApiUrl = '/rest/api/content'
# Change these based on your instance
confluenceBaseUrl = 'https://confluence.corp.bioagilytix.com/confluence'


username = 'yangyang.liu'
password = getpass("password")
confluence = Confluence(
    url='https://confluence.corp.bioagilytix.com/confluence',
    username=username,
    password=password)
```

```python
def getPageLabel(pageId):
    url = "{}{}/{}/label".format(confluenceBaseUrl, contentApiUrl, pageId)
    requestResponse = requests.get(url, auth=(username, password))
    responseList = requestResponse.json()['results']
    labelList = list()
    for label in responseList:
        labelList.append(label['name'])
    return labelList

```

```python
def getAllSpaceKey():
    responseList = confluence.get_all_spaces()['results']
    spaceList = list()
    for item in responseList:
        spaceList.append(item['key'])
    return spaceList
```

```python
def getPageIdByTitle(spaceKey, pageName):

    response = confluence.get_all_pages_from_space(spaceKey)
    for page in response:
        if page['title'] == pageName:
            return page['id']
    return 'Not Found!'
```

```python
spaceList = getAllSpaceKey()
```

```python
allNoteToFileList = list()
for spaceKey in spaceList:
    print(spaceKey)
    if getPageIdByTitle(spaceKey, "Notes to File") != "Not Found!":

        pageId = getPageIdByTitle(spaceKey, "Notes to File")
        allNoteToFileList.append(pageId)


    
```

```python
for pageId in allNoteToFileList:
    pageLabel = getPageLabel(pageId)
        
    print(pageLabel)
    if 'bpm-qc-review' in pageLabel:
        print('Label need to be removed from the page {}'.format(pageId))
```

```python
labelFound = False
for pageId in allNoteToFileList:
    pageLabel = getPageLabel(pageId)
    if len(pageLabel) > 0:
        print('The Following Label has been found: ')
        print(pageLabel)
        labelFound = True
    if 'bpm-qc-review' in pageLabel:
        print('Label need to be removed from the page {}'.format(pageId))
        labelFound = True
if labelFound == False:
    print('A total of {} Notes to File page inspected'.format(len(allNoteToFileList)))
    print('No Workflow Label has been found on any of the Notes to File Page')
```
